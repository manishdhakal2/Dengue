<h1>Dengue - Introduction </h1>
Dengue is a viral infection caused by the dengue virus, which is transmitted primarily by Aedes mosquitoes, particularly Aedes aegypti and Aedes albopictus. These mosquitoes are typically active during the day, especially at dawn and dusk. Dengue is common in tropical and subtropical regions around the world, with symptoms ranging from mild to severe.

```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import seaborn as sns
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

device="cuda" if torch.cuda.is_available() else "cpu"
```

```python
#load the data
train_x=pd.read_csv(r"train_x.csv")
train_x.head()
```

```python
#load the train_y
train_y=pd.read_csv(r"train_y.csv")
train_y.head()
```

```python
print(train_x.isna().sum())
```

```python
figure=plt.figure(figsize=(10,20))
for i in range(train_x.columns.size):
    plt.subplot(train_x.columns.size//4 +1,4,i+1)
    plt.hist(train_x.iloc[:,i],bins=80)
    plt.title(train_x.columns[i])

plt.tight_layout()
    
```

```python
correlation_matrix=train_x.corr()
plt.figure(figsize=(10, 8))  
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Features")
plt.show()
```

<b> Here, we can see that columns like avg_temp_k and avg_temp_c , specific humidity per kg and dew_point_temp ,trdr k and max_air_temp have maximum correlation . Hence we can safely remove one of these columns</b?

```python
train_x.drop("dew_point_temp_k",inplace=True,axis=1)
train_x.drop("reanalysis_tdtr_k",inplace=True,axis=1)
train_x.drop("air_temp_k",inplace=True,axis=1)
```

```python
print(train_x.isna().sum())
```

```python
figure=plt.figure(figsize=(20,20))
for i in range(train_x.columns.size):
    plt.subplot(train_x.columns.size//4 +1,4,i+1)
    plt.scatter(train_x.iloc[:,i],train_y.iloc[:,0])
    plt.xlabel(train_x.columns[i])
    plt.ylabel(train_y.columns[0])

plt.tight_layout()
```

<b> This shows that there are a few outliers in the dataset, so median can be used as imputing factor</b>

```python
for i in range(train_x.columns.size):
    train_x.iloc[:,i]=train_x.iloc[:,i].fillna(train_x.iloc[:,i].median())
```

```python
print(train_x.isna().sum())
print(train_y.isna().sum())
```

```python
X_train=np.array(train_x.iloc[:,:])
Y_train=np.array(train_y.iloc[:,0])
```

```python
#standarize the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)

#Split the data into training and testing sets
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_train,Y_train,shuffle=True,test_size=0.33)
```

```python
print(y_train.shape)
print(x_test,y_test)
```

<h2> Model Building </h2>

```python
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor(n_estimators=500)
model.fit(x_train,y_train)


```

```python
y_pred=model.predict(x_test).astype(int)
print(mean_squared_error(y_test,y_pred))
```

```python

plt.scatter(y_pred,y_test)
```

```python
class NeuralNet(nn.Module):
    def __init__(self,lr,n_features):
        super(NeuralNet,self).__init__()
        self.lr=lr
        self.model=nn.Sequential(nn.Linear(n_features,20),
                                nn.ReLU(),
                                nn.Linear(20,20),
                                nn.ReLU(),
                                nn.Linear(20,1)
                                )
        self.optimizer=torch.optim.SGD(lr=self.lr,momentum=0.9,params=self.model.parameters())

    def forward(self,x):
        return self.model(x)
```

```python
tensor_train_x=torch.tensor(x_train,device=device)
tensor_train_y=torch.tensor(y_train,device=device)
tensor_test_x=torch.tensor(x_test,device=device)
tensor_test_y=torch.tensor(y_test,device=device)
```

```python
tensor_train_x=tensor_train_x.type(torch.float)
tensor_train_y=tensor_train_y.type(torch.float)
tensor_test_x=tensor_test_x.type(torch.float)
tensor_test_y=tensor_test_y.type(torch.float)
```

```python
Netmodel=NeuralNet(0.1,10).to(device)
```

```python
#Training Loop
loss_fn=nn.MSELoss()
epoch=100
for i in range(epoch):
    Netmodel.train()
    y_pred=Netmodel.forward(tensor_train_x)
    y_pred.to(device)
    Netmodel.optimizer.zero_grad()
    loss=loss_fn(tensor_train_y,y_pred.reshape(-1))
    if(loss.item()<2.0):
        break
    loss.backward()
    Netmodel.optimizer.step()
    print(loss.item())

    
    
```
